```python
countries_gdp = pandas.read_excel("imf_gdp_estimates.xlsx")
```

```python
countries_gdp.reset_index()
```

```python
countries_gdp.info()
```

```python
countries_gdp.columns = ['country', 'subject_descriptor','units','scale', 'country_specific_notes', 'gdp_2015', 'gdp_2016', 'estimates_start_after']
```

```python
countries_gdp
```

```python
countries_gdp["gdp_2016_numeric"] = pandas.to_numeric(countries_gdp.gdp_2016, errors='coerce')
```

```python
countries_less_than_banks = countries_gdp[countries_gdp.gdp_2016_numeric < (bank_profits/1000000000)]
```

```python
countries_less_than_banks
```

```python
countries_less_than_banks.sort_values('gdp_2016_numeric', ascending=False) 
```

```python
countries_less_than_banks.sort_values('gdp_2016_numeric', ascending=True).gdp_2016_numeric.head(53).sum()
```

```python
total_countries_gdp_less_than_banks=countries_less_than_banks.sort_values('gdp_2016_numeric').head(53)
```
